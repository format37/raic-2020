```python
import pickle
import pandas as pd
```

```python
time_report = pickle.load(open("time_report_1.dat", "rb"))
rep = pd.DataFrame(time_report)
graphic = rep.plot(
    #x = rep.x,
    y=rep.columns.values[1:],
    title = sum(rep[rep.columns.values[1:]].sum()),
    figsize=(20,7),
    grid = True
)
fig = graphic.get_figure()
fig.savefig("myplot.png")
```
